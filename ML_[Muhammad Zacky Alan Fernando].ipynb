#### Perbedaan Descriptive Analysis dan Exploratory Data Analysis (EDA).
**Descriptive Analysis**

Descriptive analysis merupakan metode yang fokus pada meringkas data menjadi informasi yang mudah dipahami tanpa melakukan eksplorasi lebih lanjut atau mencari pola kompleks

- Bertujuan untuk merangkum dan menggambarkan data.

- Menjawab pertanyaan “apa yang terjadi?”.

- Menggunakan statistik dasar (mean, median, modus, persentase).

- Visualisasi sederhana seperti tabel dan grafik batang/garis.

- Biasanya digunakan untuk laporan dan ringkasan data.

**Exploratory Data Analysis (EDA)**

EDA adalah proses awal dalam analisis data yang bertujuan untuk memahami karakteristik, pola, dan hubungan antar variabel dalam dataset sebelum dilakukan analisis lanjutan atau pemodelan prediktif

- Bertujuan untuk menggali pola, hubungan, dan kejanggalan data.

- Menjawab pertanyaan “mengapa dan bagaimana?”.

- Menggunakan analisis distribusi, korelasi, dan deteksi outlier.

- Visualisasi lebih beragam (histogram, boxplot, scatter plot, heatmap).

- Digunakan sebagai tahap awal sebelum analisis lanjutan atau machine learning.

### Cleaning Data
Tujuan utama:

- Menjadikan data siap dianalisis

- Memastikan tipe data sesuai

- Meningkatkan keterbacaan & kemudahan analisis

#### Mengubah Kolom Tanggal ke Datetime

```python
day['dteday'] = pd.to_datetime(day['dteday'])
hour['dteday'] = pd.to_datetime(hour['dteday'])
```

**Insight:**
Kolom tanggal diubah ke format datetime agar dapat digunakan untuk analisis berbasis waktu seperti tren dan pola musiman.

#### Mengubah Kolom Kategorikal ke Tipe Category

```python
categorical_cols_day = ['season', 'yr', 'mnth', 'weekday', 'weathersit', 'workingday']
categorical_cols_hour = ['season', 'yr', 'mnth', 'hr', 'weekday', 'weathersit', 'workingday']

day[categorical_cols_day] = day[categorical_cols_day].astype('category')
hour[categorical_cols_hour] = hour[categorical_cols_hour].astype('category')
```

**Insight**: Variabel waktu dan kondisi tertentu diubah ke tipe data kategori untuk meningkatkan kejelasan struktur data dan mendukung analisis kategorikal.

#### Menghapus Kolom yang Tidak Digunakan

```python
day_df_clean = day.drop(columns=['casual', 'registered'])
hour_df_clean = hour.drop(columns=['casual', 'registered'])
```

**Insight:** kolom casual dan registered dihapus karena secara langsung membentuk nilai cnt dan berpotensi menyebabkan data leakage pada pemodelan machine learning nantinya.

```python
day_df_clean.info()
hour_df_clean.info()
```

Dapat disimpulkan pada tahap data cleaning:
- Tipe data sudah sesuai
- Tidak ada missing & duplikasi
- Data leakage dicegah
- Dataset siap untuk EDA

## Exploratory Data Analysis (EDA)

```python
day.describe()
```

**insight**
- cnt (total peminjaman) memiliki variasi yang cukup besar:  indikasi dipengaruhi faktor eksternal

- temp dan atemp memiliki rentang nilai yang relatif stabil

- windspeed cenderung rendah pada sebagian besar observasi

```python
import matplotlib.pyplot as plt

plt.figure()
plt.hist(day_df_clean['cnt'], bins=30)
plt.xlabel('Total Rentals (cnt)')
plt.ylabel('Frequency')
plt.title('Distribution of Bike Rentals')
plt.show()
```

Distribusi jumlah penyewaan sepeda menunjukkan bahwa sebagian besar hari memiliki jumlah penyewaan pada level menengah, dengan beberapa hari memiliki permintaan yang sangat tinggi.

```python
day_df_clean.groupby('season', observed = True)['cnt'].mean().plot(kind='bar')
plt.xlabel('Season')
plt.ylabel('Average Rentals')
plt.title('Average Bike Rentals by Season')
plt.show()
```

**Insight:**
- Musim tertentu (summer & fall) memiliki rata-rata penyewaan lebih tinggi

- Faktor cuaca dan aktivitas luar ruang sangat berpengaruh terhadap jumlah peminjaman

```python
day_df_clean.groupby('weathersit', observed=True)['cnt'].mean().plot(kind='line')
plt.xlabel('Weather Situation')
plt.ylabel('Average Bike Rentals')
plt.title('Average Bike Rentals by Weather Condition')
plt.show()
```

- Cuaca cerah memiliki rata-rata peminjaman tertinggi

- Cuaca buruk (hujan/salju) menurunkan jumlah peminjaman secara signifikan

```python
plt.figure()
plt.scatter(day_df_clean['temp'], day_df_clean['cnt'])
plt.xlabel('Temperature')
plt.ylabel('Total Rentals')
plt.title('Temperature vs Bike Rentals')
plt.show()
```

- Hubungan positif

- Suhu yang lebih hangat → peminjaman meningkat

```python
plt.figure()
plt.scatter(day_df_clean['hum'], day_df_clean['cnt'])
plt.xlabel('Humidity')
plt.ylabel('Total Rentals')
plt.title('Humidity vs Bike Rentals')
plt.show()
```

- Hubungan negatif

- Kelembapan tinggi → peminjaman menurun

```python
day_df_clean.groupby('workingday', observed=True)['cnt'].mean().plot(kind='bar')
plt.xlabel('Working Day')
plt.ylabel('Average Rentals')
plt.title('Bike Rentals: Working Day vs Non-Working Day')
plt.show()
```

- Hari kerja cenderung memiliki rata-rata peminjaman lebih tinggi

- Mengindikasikan penggunaan sepeda untuk aktivitas commuting

```python
day_df_clean.boxplot(column='cnt', by='workingday')
plt.xlabel('Working Day (0 = Libur, 1 = Kerja)')
plt.ylabel('Total Rentals')
plt.title('Distribution of Bike Rentals by Working Day')
plt.suptitle('')
plt.show()
```

- Hari kerja (1) memiliki median peminjaman sedikit lebih tinggi dibanding hari libur (0)

    Artinya, rata-rata aktivitas peminjaman lebih konsisten di hari kerja
- Hari kerja menunjukkan variasi yang lebih besar

    Rentang data lebih lebar pada hari kerja dan 
    menunjukkan perbedaan aktivitas antar hari kerja (misalnya Senin vs Jumat)

- hari libur lebih stabil walaupun ada beberapa hari dengan peminjaman tinggi

```python
corr_matrix = day_df_clean[['temp', 'atemp', 'hum', 'windspeed', 'workingday', 'cnt']].corr()
plt.figure(figsize=(8, 5))
sns.heatmap(corr_matrix,
            annot=True,
            linewidths=0.5,
            fmt= ".2f",
            cmap="YlGnBu");
```

Faktor suhu memiliki korelasi positif terhadap jumlah penyewaan, sementara kelembapan dan kecepatan angin cenderung menurunkan permintaan sepeda.

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana pengaruh Kondisi Lingkungan terhadap Jumlah Peminjaman Sepeda?
Varibel yang akan digunakan:
- weathersit → kondisi cuaca
- temp → suhu
- hum → kelembapan
- cnt → jumlah peminjaman sepeda

```python
day_df_clean.groupby('weathersit', observed=True)['cnt'].mean().plot(kind='bar')
plt.xlabel('Weather Situation')
plt.ylabel('Average Bike Rentals')
plt.title('Average Bike Rentals by Weather Condition')
plt.show()

plt.figure()
plt.scatter(day_df_clean['temp'], day_df_clean['cnt'])
plt.xlabel('Temperature')
plt.ylabel('Total Rentals')
plt.title('Temperature vs Bike Rentals')
plt.show()

plt.figure()
plt.scatter(day_df_clean['hum'], day_df_clean['cnt'])
plt.xlabel('Humidity')
plt.ylabel('Total Rentals')
plt.title('Humidity vs Bike Rentals')
plt.show()
```


**Insight**: 

Jumlah peminjaman sepeda tertinggi terjadi pada kondisi cuaca cerah. Semakin buruk kondisi cuaca, rata-rata peminjaman sepeda semakin menurun, menunjukkan bahwa cuaca merupakan faktor utama yang memengaruhi permintaan sepeda.

Terdapat hubungan positif antara suhu dan jumlah peminjaman sepeda, di mana peminjaman cenderung meningkat pada suhu yang lebih hangat.

Kelembapan yang tinggi cenderung menurunkan jumlah peminjaman sepeda, karena kondisi lingkungan menjadi kurang nyaman untuk beraktivitas di luar ruangan. Disisi lain kelembapan yang terlalu rendah juga menurunkan jumlah peminjaman sepeda.

### Pertanyaan 2: Bagaimana perbedaan jumlah peminjaman sepeda antara hari kerja dan hari libur?
Variabel yang Digunakan:
- workingday → 1 (hari kerja), 0 (hari libur)

- cnt → jumlah peminjaman sepeda


```python
day_df_clean.boxplot(column='cnt', by='workingday')
plt.xlabel('Working Day (0 = Libur, 1 = Kerja)')
plt.ylabel('Total Rentals')
plt.title('Distribution of Bike Rentals by Working Day')
plt.suptitle('')
plt.show()

day_df_clean.groupby('workingday', observed=True)['cnt'].mean().plot(kind='bar')
plt.xlabel('Working Day (0 = Libur, 1 = Kerja)')
plt.ylabel('Average Bike Rentals')
plt.title('Average Bike Rentals: Working Day vs Holiday')
plt.show()
```

**insight**:

Rata-rata peminjaman sepeda pada hari kerja lebih tinggi dibandingkan hari libur, yang menunjukkan bahwa sepeda banyak digunakan sebagai sarana transportasi harian.

Distribusi peminjaman sepeda pada hari kerja menunjukkan variasi yang lebih besar dibandingkan hari libur, mengindikasikan fluktuasi permintaan yang lebih tinggi pada hari kerja.
